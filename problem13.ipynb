```julia
import Random.Random
#I *think* this is the solution of problem 13. I'm not quite sure
#because I've still got doubts about calculating the computation time
#(but it works at least)

function findK(v, k)
    if k<=0
        return "Failure"
    end
    n=length(v)
    if k==1   #trivial case: if k=1, search for largest value
        a=v[1]
        for i =2:n
            if a<v[i]
                a=v[i]
            end
        end
        return a
    elseif k==n #trivial case 2: if k=n, search for smallest value
        a=v[1]
        for i=2:n
            if a>v[i]
                a=v[i]
            end
        end
        return a
    end
    j=abs(rand(Int))%n+1 #I choose a random element and build two arrays like in quicksort
    left=[x for x in v[1:n] if x<=v[j]]
    right=[x for x in v[1:n] if x>v[j]]
    m=length(right)
    if m<k #if the right array is smaller than k, I repeat the algorithm with left array, looking for (k-m)th element
        return findK(left, k-m)
    else        
        return findK(right, k) #if right array is bigger or equal than k, it must contain the kth element and I repeat the algorithm with it
    end
end
            
```
